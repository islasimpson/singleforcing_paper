```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from math import nan

from CASutils import averaging_utils as avg

import sys
```

```python
plotpath='/project/cas/islas/python_plots/singleforcing_paper/figures/supplemental/'
```

```python
aer2 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/FORCINGS/AAER.nc")
aer1 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/FORCINGS/AAER_LENS1.nc")
```

```python
aer2_so2 = aer2.ene_ind_so2 + aer2.ag_so2 + aer2.restran_so2 + aer2.ship_so2
aer2_so4 = aer2.ene_ind_so4 + aer2.ag_so4 + aer2.restran_so4 + aer2.ship_so4
```

```python
fig = plt.figure(figsize=(16,16))

ax = fig.add_axes([0.0,0.8,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_title('(a) AAER black carbon', fontsize=16)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
ax.set_ylim(0,10)
ax.set_yticks([0,2,4,6,8,10])
ax.set_yticklabels(['0','2','4','6','8','10'], fontsize=14)
ax.set_ylabel('Emissions (Tg/year)', fontsize=14)

ax.plot(aer2.year, aer2.anthro_bc, color='maroon', linewidth=2, label='LENS2')
ax.plot(aer1.year, aer1.anthro_bc, color='teal', linewidth=2, label='LENS1')

ax.legend()



ax = fig.add_axes([0.32,0.8,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_title('(b) AAER SO4', fontsize=16)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
ax.set_ylabel('Emissions (Tg/year)', fontsize=14)
ax.set_ylim(0,6.5)
ax.set_yticks([0,1,2,3,4,5,6])
ax.set_yticklabels(['0','1','2','3','4','5','6'], fontsize=14)

ax.plot(aer2.year, aer2_so4, color='maroon', linewidth=2, label='LENS2-SF')
ax.plot(aer1.year, aer1.anthro_so4, color='teal', linewidth=2, label='LENS1')

ax.legend()


ax = fig.add_axes([0.64,0.8,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_title('(c) AAER SO2', fontsize=16)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
ax.set_ylim(0,140)
ax.set_yticks([0,20,40,60,80,100,120,140])
ax.set_yticklabels(['0','20','40','60','80','100','120','140'], fontsize=14)
ax.set_ylabel('Emissions (Tg/year)', fontsize=14)


ax.plot(aer2.year, aer2_so2, color='maroon', linewidth=2, label='LENS2-SF')
ax.plot(aer1.year, aer1.anthro_so2, color='teal', linewidth=2, label='LENS1')

ax.legend()



ax = fig.add_axes([0.16,0.57,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_title('(d) AAER POM', fontsize=16)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
#ax.set_ylim(3,40)
ax.set_yticklabels([5,10,15,20,25,30,35,40])
ax.set_yticklabels(['5','10','15','20','25','30','35','40'], fontsize=14)

ax.set_ylabel('Emissions (Tg/year)', fontsize=14)

ax.plot(aer2.year, aer2.pom, color='maroon', linewidth=2, label='LENS2-SF')
ax.plot(aer1.year, aer1.anthro_pom, color='teal', linewidth=2, label='LENS1')

ax.legend(loc='upper left')


ax = fig.add_axes([0.48,0.57,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_title('(e) AAER SOAG', fontsize=16)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)

ax.set_ylabel('Emissions (Tg/year)', fontsize=14)

ax.plot(aer2.year, aer2.soag, color='maroon', linewidth=2, label='LENS2-SF')
#ax.plot(aer1.year, aer1.anthro_soag, color='teal', linewidth=2, label='LENS1')

ax.legend(loc='upper left')

fig.savefig(plotpath+'figS3.png', bbox_inches='tight', facecolor='white')
fig.savefig(plotpath+'figS3.pdf', bbox_inches='tight', facecolor='white')
```
