### DJFM Sea ice thickness, 220E-360E, 70N-90N

```python
dat = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/ocn_wgts/b.e21.BSSP370smbb.f09_g17.LE2-1301.018.cice.h.hi.205501-206412.nc")
wgts = dat.tarea
wgts = xr.where( (dat.TLON > 220) & (dat.TLON < 360), wgts, 0)
wgts = xr.where( (dat.TLAT > 70) & (dat.TLAT < 90), wgts, 0)
```

```python
xaaer1 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS1-SF/XAER_sicthick_djfm.nc")
xaaer1 = xaaer1.mean('M').hi
xaaer1['time'] = xaaer1.time.dt.year
```

```python
xaaer2 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/CESM2-XAAER/xAER_sicthick_djfm.nc")
xaaer2 = xaaer2.mean('M').hi
xaaer2['time'] = xaaer2.time.dt.year
```

```python
xaaer1_base = xaaer1.sel(time=slice(1920,1940)).mean('time')
xaaer2_base = xaaer2.sel(time=slice(1920,1940)).mean('time')
```

```python
xaaer1_anoms = xaaer1 - xaaer1_base
xaaer2_anoms = xaaer2 - xaaer2_base
```

```python
xaaer1_w = xaaer1_anoms.weighted(wgts.fillna(0))
xaaer1_hi_avg = xaaer1_w.mean(("ni","nj"))

xaaer2_w = xaaer2_anoms.weighted(wgts.fillna(0))
xaaer2_hi_avg = xaaer2_w.mean(("ni","nj"))
```

```python
xaaer1_hi_avg_21y = xaaer1_hi_avg.rolling(time=21, min_periods=21, center='True').mean('time').dropna('time')
xaaer2_hi_avg_21y = xaaer2_hi_avg.rolling(time=21, min_periods=21, center='True').mean('time').dropna('time')
```

### March Labrador sea MLD and DJFM surface heat and water fluxes over the lab sea, 300E-315E, 53N-65N

```python
# Read in the weights and set the values for the region outside the lab sea to zero
wgts=xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/ocn_wgts/b.e21.BSSP370smbb.f09_g17.LE2-1301.018.pop.h.HMXL.208501-209412.nc").TAREA
wgts = xr.where( (wgts.TLONG > 300) & (wgts.TLONG < 315), wgts, 0)
wgts = xr.where( (wgts.TLAT > 53) & (wgts.TLAT < 65), wgts, 0)
```

```python
xaaer1 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS1-SF/XAER_March_HMXL.nc")
xaaer1 = xaaer1.mean('M').HMXL

xaaer1_sfwf = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS1-SF/xAER_SFWF_DJFM.nc")
xaaer1_sfwf = xaaer1_sfwf.mean('M').SFWF

xaaer1_qflux = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS1-SF/xAER_QFLUX_DJFM.nc")
xaaer1_qflux = xaaer1_qflux.mean('M').QFLUX

xaaer1_shf = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS1-SF/xAER_SHF_DJFM.nc")
xaaer1_shf = xaaer1_shf.mean('M').SHF
```

```python
xaaer2 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/CESM2-XAAER/xAER_March_HMXL.nc")
xaaer2 = xaaer2.mean('M').HMXL

xaaer2_sfwf = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/CESM2-XAAER/xAER_SFWF_DJFM.nc")
xaaer2_sfwf = xaaer2_sfwf.mean('M').SFWF

xaaer2_qflux = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/CESM2-XAAER/xAER_QFLUX_DJFM.nc")
xaaer2_qflux = xaaer2_qflux.mean('M').QFLUX

xaaer2_shf = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/CESM2-XAAER/xAER_SHF_DJFM.nc")
xaaer2_shf = xaaer2_shf.mean('M').SHF

# Latent heat of fusion in units of J/kg
lf = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/ocn_wgts/b.e21.BSSP370smbb.f09_g17.LE2-1301.018.pop.h.HMXL.208501-209412.nc")
lf = (10**(-7))*1000.*lf.latent_heat_fusion.values
```

```python
xaaer1['time'] = xaaer1.time.dt.year
xaaer1_sfwf['time'] = xaaer1_sfwf.time.dt.year
xaaer1_qflux['time'] = xaaer1_qflux.time.dt.year
xaaer1_shf['time'] = xaaer1_shf.time.dt.year

xaaer2['time'] = xaaer2.time.dt.year
xaaer2_sfwf['time'] = xaaer2_sfwf.time.dt.year
xaaer2_qflux['time'] = xaaer2_qflux.time.dt.year
xaaer2_shf['time'] = xaaer2_shf.time.dt.year
```

```python
xaaer1 = xaaer1.sel(time=slice(xaaer1_sfwf.time.min('time').values, xaaer1_sfwf.time.max('time').values))
xaaer2 = xaaer2.sel(time=slice(xaaer2_sfwf.time.min('time').values, xaaer2_sfwf.time.max('time').values))
```

```python
xaaer1 = xaaer1.sel(time=slice(1921,2050)) ; xaaer2 = xaaer2.sel(time=slice(1921,2050))
xaaer1_sfwf = xaaer1_sfwf.sel(time=slice(1921,2050)) ; xaaer2_sfwf = xaaer2_sfwf.sel(time=slice(1921,2050))
xaaer1_qflux = xaaer1_qflux.sel(time=slice(1921,2050)) ; xaaer2_qflux = xaaer2_qflux.sel(time=slice(1921,2050))
xaaer1_shf = xaaer1_shf.sel(time=slice(1921,2050)) ; xaaer2_shf = xaaer2_shf.sel(time=slice(1921,2050))
```

```python
xaaer1_lab_w = xaaer1.weighted(wgts.fillna(0))
xaaer1_lab_m = xaaer1_lab_w.mean(("nlon","nlat"))

xaaer1_sfwf_w = xaaer1_sfwf.weighted(wgts.fillna(0))
xaaer1_sfwf_m = xaaer1_sfwf_w.mean(("nlon","nlat"))

xaaer1_qflux_w = xaaer1_qflux.weighted(wgts.fillna(0))
xaaer1_qflux_m = xaaer1_qflux_w.mean(("nlon","nlat"))

xaaer1_shf_w = xaaer1_shf.weighted(wgts.fillna(0))
xaaer1_shf_m = xaaer1_shf_w.mean(("nlon","nlat"))

xaaer2_lab_w = xaaer2.weighted(wgts.fillna(0))
xaaer2_lab_m = xaaer2_lab_w.mean(("nlon","nlat"))

xaaer2_sfwf_w = xaaer2_sfwf.weighted(wgts.fillna(0))
xaaer2_sfwf_m = xaaer2_sfwf_w.mean(("nlon","nlat"))

xaaer2_qflux_w = xaaer2_qflux.weighted(wgts.fillna(0))
xaaer2_qflux_m = xaaer2_qflux_w.mean(("nlon","nlat"))

xaaer2_shf_w = xaaer2_shf.weighted(wgts.fillna(0))
xaaer2_shf_m = xaaer2_shf_w.mean(("nlon","nlat"))
```

```python
water_flux_xaaer1 = xaaer1_sfwf_m - xaaer1_qflux_m/lf
heat_flux_xaaer1 = xaaer1_shf_m + xaaer1_qflux_m

water_flux_xaaer2 = xaaer2_sfwf_m - xaaer2_qflux_m/lf
heat_flux_xaaer2 = xaaer2_shf_m + xaaer2_qflux_m
```

```python
mld_anoms_xaaer1 = xaaer1_lab_m - xaaer1_lab_m.sel(time=slice(1920,1940)).mean('time')
mld_anoms_xaaer2 = xaaer2_lab_m - xaaer2_lab_m.sel(time=slice(1920,1940)).mean('time')

water_flux_xaaer1 = water_flux_xaaer1 - water_flux_xaaer1.sel(time=slice(1920,1940)).mean('time')
water_flux_xaaer2 = water_flux_xaaer2 - water_flux_xaaer2.sel(time=slice(1920,1940)).mean('time')

heat_flux_xaaer1 = heat_flux_xaaer1 - heat_flux_xaaer1.sel(time=slice(1920,1940)).mean('time')
heat_flux_xaaer2 = heat_flux_xaaer2 - heat_flux_xaaer2.sel(time=slice(1920,1940)).mean('time')
```

```python
mld_anoms_xaaer1_21y = mld_anoms_xaaer1.rolling(time=21, min_periods=21, center='True').mean('time').dropna('time')
mld_anoms_xaaer2_21y = mld_anoms_xaaer2.rolling(time=21, min_periods=21, center='True').mean('time').dropna('time')

water_flux_xaaer1_21y = water_flux_xaaer1.rolling(time=21, min_periods=21, center='True').mean('time').dropna('time')
water_flux_xaaer2_21y = water_flux_xaaer2.rolling(time=21, min_periods=21, center='True').mean('time').dropna('time')

heat_flux_xaaer1_21y = heat_flux_xaaer1.rolling(time=21, min_periods=21, center='True').mean('time').dropna('time')
heat_flux_xaaer2_21y = heat_flux_xaaer2.rolling(time=21, min_periods=21, center='True').mean('time').dropna('time')
```
