```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
plotpath="/project/cas/islas/python_plots/singleforcing_paper/figures/supplemental/"
```

```python
ghgs2 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/FORCINGS/GHGs.nc")
ghgs1 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/FORCINGS/GHGs_LENS1.nc")
```

```python
fig = plt.figure(figsize=(16,16))

ax = fig.add_axes([0.0,0.8,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
ax.set_ylim(280,550)
ax.set_yticks([300,350,400,450,500,550])
ax.set_yticklabels(['300','350','400','450','500','550'], fontsize=14)
ax.set_ylabel('CO$_{2}$ (ppmv)', fontsize=14)
ax.set_title('(a) CO$_{2}$', fontsize=16)

ax.plot(ghgs1.year, 1e6*ghgs1.co2vmr, color='teal', linewidth=3, label='LENS1')
ax.plot(ghgs2.year, 1e6*ghgs2.co2vmr, color='maroon', linewidth=3, linestyle='dashed', label='LENS2')
ax.legend()

ax = fig.add_axes([0.32,0.8,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
ax.set_ylim(0.75,2.5)
ax.set_yticks([0.8,1.2,1.6,2.0,2.4])
ax.set_yticklabels(['0.8','1.2','1.6','2.0','2.4'], fontsize=12)
ax.set_title('(b) CH$_{4}$', fontsize=16)
ax.set_ylabel('CH$_{4}$ (ppmv)', fontsize=14)

ax.plot(ghgs1.year, 1e6*ghgs1.ch4vmr, color='teal', linewidth=3, label='LENS1')
ax.plot(ghgs2.year, 1e6*ghgs2.ch4vmr, color='maroon', linewidth=3, linestyle='dashed', label='LENS2') 
ax.legend()

ax = fig.add_axes([0.64,0.8,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
ax.set_ylim(0.26,0.38)
ax.set_yticks([0.26,0.28,0.3,0.32,0.34,0.36,0.38])
ax.set_yticklabels(['0.26','0.28','0.30','0.32','0.34','0.36','0.38'], fontsize=14)
ax.set_title('(c) N$_{2}$O', fontsize=16)
ax.set_ylabel('N$_{2}$O (ppmv)', fontsize=14)

ax.plot(ghgs1.year, 1e6*ghgs1.n2ovmr, color='teal', linewidth=3, label='LENS1')
ax.plot(ghgs2.year, 1e6*ghgs2.n2ovmr, color='maroon', linewidth=3, linestyle='dashed', label='LENS2')
ax.legend()

ax = fig.add_axes([0.16,0.57,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
ax.set_ylim(0,0.0011)
ax.set_yticks([0,0.0002,0.0004,0.0006,0.0008,0.0010])
ax.set_yticklabels(['0','2','4','6','8','10'], fontsize=14)
ax.set_ylabel('F11 ($\\times$10$^{-4}$ ppmv)', fontsize=14)
ax.set_title('(d) F11', fontsize=16)

ax.plot(ghgs1.year, 1e6*ghgs1.f11vmr, color='teal', linewidth=3, label='LENS1')
ax.plot(ghgs2.year, 1e6*ghgs2.f11vmr, color='maroon', linewidth=3, linestyle='dashed', label='LENS2')
ax.legend()

ax = fig.add_axes([0.48,0.57,0.25,0.17])
ax.set_xlim(1850,2050)
ax.set_xticks([1850,1900,1950,2000,2050])
ax.set_xticklabels(['1850','1900','1950','2000','2050'], fontsize=14)
ax.set_ylim(0,0.0006)
ax.set_yticks([0,0.0001,0.0002,0.0003,0.0004,0.0005,0.0006])
ax.set_yticklabels(['0','1','2','3','4','5','6'], fontsize=14)
ax.set_ylabel('F12 ($\\times$10$^{-4}$ ppmv)', fontsize=14)
ax.set_title('(e) F12', fontsize=16)

ax.plot(ghgs1.year, 1e6*ghgs1.f12vmr, color='teal', linewidth=3, label='LENS1')
ax.plot(ghgs2.year, 1e6*ghgs2.f12vmr, color='maroon', linewidth=3, linestyle='dashed', label='LENS2')
ax.legend()

fig.savefig(plotpath+'figS2.png', bbox_inches='tight', facecolor='white')
fig.savefig(plotpath+'figS2.pdf', bbox_inches='tight', facecolor='white')
```
