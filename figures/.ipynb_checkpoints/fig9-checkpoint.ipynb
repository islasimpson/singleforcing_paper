```python
def calc21ymean(dat):
    datm = dat.rolling(year=21, min_periods=21, center='True').mean('year').dropna('year')
    return datm
```

### Read in AOD

```python
bootaod1 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/BOOTSTRAP/fig9/CESM1_AODVIS_bootstrap.nc")
bootaod2 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/BOOTSTRAP/fig9/CESM2_AODVIS_bootstrap.nc")
```

```python
aaer2 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS2-SF/AAER_AODVIS_am.nc")
lens2 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS2/LENS2_AODVIS_am.nc")
xaer2 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/CESM2-XAAER/xAER_AODVIS_am.nc")
```

```python
aaer1 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/CESM1-AAER/AAER_AODVIS_am.nc")
lens1 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS1/LENS1_AODVIS_am.nc")
xaer1 = xr.open_dataset("/project/cas/islas/python_savs/singleforcing_paper/DATA_SORT/LENS1-SF/XAER_AODVIS_am.nc")
```

```python
aaer2base = aaer2.sel(year=slice(1920,1940)).mean('year')
lens2base = lens2.sel(year=slice(1920,1940)).mean('year')
xaer2base = xaer2.sel(year=slice(1920,1940)).mean('year')

aaer1base = aaer1.sel(year=slice(1920,1940)).mean('year')
lens1base = lens1.sel(year=slice(1920,1940)).mean('year')
xaer1base = xaer1.sel(year=slice(1920,1940)).mean('year')
```

```python
aaer2anoms = aaer2 - aaer2base
lens2anoms = lens2 - lens2base
xaer2anoms = xaer2 - xaer2base

aaer1anoms = aaer1 - aaer1base
lens1anoms = lens1 - lens1base
xaer1anoms = xaer1 - xaer1base
```

```python
aaer2anoms = aaer2anoms.mean('M')
lens2anoms = lens2anoms.mean('M')
xaer2anoms = xaer2anoms.mean('M')

aaer1anoms = aaer1anoms.mean('M')
lens1anoms = lens1anoms.mean('M')
xaer1anoms = xaer1anoms.mean('M')
```

```python
aaer2_aod_sm = avg.cosweightlonlat(aaer2anoms, 0, 360, 50, 90)
lens2_aod_sm = avg.cosweightlonlat(lens2anoms, 0, 360, 50, 90)
xaer2_aod_sm = avg.cosweightlonlat(xaer2anoms, 0, 360, 50, 90)

aaer1_aod_sm = avg.cosweightlonlat(aaer1anoms, 0, 360, 50, 90)
lens1_aod_sm = avg.cosweightlonlat(lens1anoms, 0, 360, 50, 90)
xaer1_aod_sm = avg.cosweightlonlat(xaer1anoms, 0, 360, 50, 90)
```

```python
aaer2_aod_sm = calc21ymean(aaer2_aod_sm)
lens2_aod_sm = calc21ymean(lens2_aod_sm)
xaer2_aod_sm = calc21ymean(xaer2_aod_sm)

aaer1_aod_sm = calc21ymean(aaer1_aod_sm)
lens1_aod_sm = calc21ymean(lens1_aod_sm)
xaer1_aod_sm = calc21ymean(xaer1_aod_sm)
```

### Calculate anomalies

```python
lens1base = lens1_amoc.sel(year=slice(1920,1940)).mean('year')
xaer1base = xaer1_amoc.sel(year=slice(1920,1940)).mean('year')
aaer1base = aaer1_amoc.sel(year=slice(1920,1940)).mean('year')
xghg1base = xghg1_amoc.sel(year=slice(1920,1940)).mean('year')
xbmb1base = xbmb1_amoc.sel(year=slice(1920,1940)).mean('year')

lens1anoms = lens1_amoc - lens1base
xaer1anoms = xaer1_amoc - xaer1base
aaer1anoms = aaer1_amoc - aaer1base
xghg1anoms = xghg1_amoc - xghg1base
xbmb1anoms = xbmb1_amoc - xbmb1base

lens1anoms = lens1anoms.mean('M')
xaer1anoms = xaer1anoms.mean('M')
aaer1anoms = aaer1anoms.mean('M')
xghg1anoms = xghg1anoms.mean('M')
xbmb1anoms = xbmb1anoms.mean('M')

aaer1xway = lens1anoms - xaer1anoms
ghg1xway = lens1anoms - xghg1anoms
bmb1xway = lens1anoms - xbmb1anoms
```

```python
aaer2base = aaer2_amoc.sel(year=slice(1920,1940)).mean('year')
lens2base = lens2_amoc.sel(year=slice(1920,1940)).mean('year')
xaer2base = xaer2_amoc.sel(year=slice(1920,1940)).mean('year')
ghg2base = ghg2_amoc.sel(year=slice(1920,1940)).mean('year')
bmb2base = bmb2_amoc.sel(year=slice(1920,1940)).mean('year')
ee2base = ee2_amoc.sel(year=slice(1920,1940)).mean('year')

aaer2anoms = aaer2_amoc - aaer2base
lens2anoms = lens2_amoc - lens2base
xaer2anoms = xaer2_amoc - xaer2base
ghg2anoms = ghg2_amoc - ghg2base
bmb2anoms = bmb2_amoc - bmb2base
ee2anoms = ee2_amoc - ee2base

aaer2anoms = aaer2anoms.mean('M')
lens2anoms = lens2anoms.mean('M')
xaer2anoms = xaer2anoms.mean('M')
ghg2anoms = ghg2anoms.mean('M')
bmb2anoms = bmb2anoms.mean('M')
ee2anoms = ee2anoms.mean('M')

aaer2xway = lens2anoms - xaer2anoms
```

```python
lens1sm = calc21ymean(lens1anoms)
xaer1sm = calc21ymean(xaer1anoms)
aaer1sm = calc21ymean(aaer1anoms)
xghg1sm = calc21ymean(xghg1anoms)
aaer1xwaysm = calc21ymean(aaer1xway)
ghg1xwaysm = calc21ymean(ghg1xway)
bmb1xwaysm = calc21ymean(bmb1xway)
```

```python
aaer2sm = calc21ymean(aaer2anoms)
lens2sm = calc21ymean(lens2anoms)
xaer2sm = calc21ymean(xaer2anoms)
ghg2sm = calc21ymean(ghg2anoms)
bmb2sm = calc21ymean(bmb2anoms)
ee2sm = calc21ymean(ee2anoms)
aaer2xwaysm = calc21ymean(aaer2xway)
```
